```python
class Man:
    def __init__(self,name):
        self.name = name
        print("start clas Man")
        
    def hello(self):
        print("Hello " + self.name + "!")
        
    def goodbye(self):
        print("Good-bye " + self.name + "!")

m = Man("dansay")
m.hello()
m.goodbye()
```

넘파이 배열 np.array 는 list를 인수로 받아 numpy.ndarray 로 반환해 준다.
넘파이 배열은 N 차원 배열을 작성할 수있다.
1차원 배열 : 벡터
2차원 배열 : 행렬
3차원 이상의 배열 : 다차원 배열

텐서 tensor : 벡터와 행렬을 일반화 한 것.

```python
x=np.array([1.0,2.0,3.0])
print(x)
print(type(x))
```

```python
x= np.array([1.0,2.0,3.0])
y= np.array([2.0,4.0,6.0])
x+y
x*y
x*2
```

```python
A=np.array([[1,2],[3,4]])
print(A)
A.shape
A.dtype
B=np.array([[10,10],[20,20]])
C=np.array([100,100])
A+B
A*B
A*C
```

```python
X=np.array([[51,55],[14,19],[0,4]])
print(X)
X[0]
X[0][1]

for dansay in X[:2]:
    print(dansay)
```

```python
X=X.flatten()
print(X)
X[np.array([0,2,4])]
X>15
X1 = X[X>15]  #X1 : bool배열 : 원소 각각에 부등호 연산을 수행하여 생성
print(X1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,6,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,10,0.1)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1,label="sin graph")
plt.plot(x,y2,label="cos graph", linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos graph')
plt.legend()
plt.show
```
